```python
import numpy as np
import scanpy as sc
import seaborn as sns
from scipy.stats import median_abs_deviation
```

# get atac data 

```python
atac = sc.read("../data/SEAAD_MTG_ATACseq_final-nuclei.2024-12-06.h5ad", backed = "r")
```

```python
# subset to microglia
mask = atac.obs["Subclass"].str.contains("Microglia-PVM")
atac = atac[mask]
atac
```

# get rna data

```python
rna = sc.read("../data/Microglia-PVM - MTG: Seattle Alzheimer's Disease Atlas (SEA-AD).h5ad", backed = "r")
```

# get intersection of atac and rna cells

```python
intersection = rna.obs_names.intersection(atac.obs_names)
```

```python
rna_subset = rna[intersection]
atac_subset = atac.to_memory()[intersection].copy()
```

```python
atac_subset.write("../data/atac_subset.h5ad")
rna_subset.write("../data/rna_subset.h5ad")
```
